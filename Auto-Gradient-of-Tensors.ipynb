```python
# Define a tensor object and also specify that it requires a gradient to be calculated:

import torch
x = torch.tensor([[2., -1.], [1., 1.]], requires_grad=True)
print(x)

# The requires_grad parameter specifies that the gradient is to be calculated for the tensor object.
```

```python
# Calculate the output of Gradient for Tensor Object
out = x.pow(2).sum()
```

```python
# The gradient of a value can be calculated by calling the backward() method to the value. 
# Calculate the gradient â€“ change in out (output) for a small change in x (input)

out.backward()
```

```python
x.grad
```

The gradients obtained previously match with the intuitive gradient
values (which are two times that of the value of x).
